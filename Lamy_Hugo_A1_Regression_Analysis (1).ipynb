Overview of OLS 

```python
import statsmodels.formula.api as smf

# Building a base model

# Instantiating a model type
lm_practice = smf.ols(formula = "REVENUE ~ TOTAL_MEALS_ORDERED",
                      data = ACDD)

# Fitting the data to the blueprint
results = lm_practice.fit()

# Printing a summary of the results
print(results.summary())
```


The term "REVENUE" serves as the dependent variable in this context. (Y)
Based on the independent variables, the model seeks to forecast the value of the dependent variable. 
R-squared: A result of 0.361 indicates that the independent variables in the model can account for 36.1% of the variance in REVENUE. 
R-squared adjusted for the number of independent variables in the model is referred to as the adj. R-squared. 
A greater number denotes a better model fit. which is NOT the situation in our circumstance. 

This regression analysis's methodology, known as "Least Squares," was applied.

The F-statistic is used to examine if the independent variables in the model are all equal to zero, which is the null hypothesis. 
A high value denotes a large contribution of the independent factors to the variance in the dependent variable.

Prob (F-statistic): This is the p-value associated with the F-statistic, which measures the significance of the independent 
variables in the model. A low p-value indicates that the independent variables are significant.
This is the model's log-likelihood, which gauges the model's degree of goodness of fit. 
The Akaike Information Criteria, or AIC, is a gauge of a model's relative excellence in 
comparison to other models. Better model fit is indicated by a lower AIC value.


Df Residuals: The number of observations minus the number of independent variables in the model corresponds 
to the degrees of freedom for the residuals.

Df Model: This refers to the model's degrees of freedom, or the number of independent variables. 
Covariance Type: In this situation, a "nonrobust" covariance matrix was employed in the regression analysis.

Coef: This is the independent variable "TOTAL MEALS ORDEREDcoefficient. "'s It indicates the change in "REVENUE," 
the dependent variable, 
for a change in "TOTAL MEALS ORDERED," expressed as a unit change.

Standard Error: This is a measure of the coefficient estimate's variability. It is the standard error of the coefficient. 
The t-statistic for the coefficient, or t, examines whether the coefficient is equal to zero, the null hypothesis.

The p-value for the t-statistic, which assesses the importance of the coefficient, is P>|t|. 
When the p-value is low, the coefficient is considered significant. 
The 95% confidence interval of the coefficient is [0.025, 0.975], and it gives an idea of the possible range of values for the coefficient.
